```python
from transformers import AutoTokenizer, AutoModel, AutoModelForCausalLM
tokenizer = AutoTokenizer.from_pretrained("silk-road/Haruhi-Zero-6B-0.1", trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained("silk-road/Haruhi-Zero-6B-0.1", device_map="auto", trust_remote_code=True)
model = model.eval()
```

```python
role_dict = {"System":"user","User":"user","AI":"bot"}
```

正常角色的测试

```python
from ChatHaruhi import ChatHaruhi

chatbot = ChatHaruhi( role_name = 'haruhi', max_len_story = 1000, llm = "foo" )

messages = chatbot.generate_messages(role='阿虚', text = '我看新一年的棒球比赛要开始了！我们要去参加吗？')

for msg in messages:
    msg['role'] = role_dict[msg['role']]

input_ids = tokenizer.apply_chat_template(conversation=messages, tokenize=True, add_generation_prompt=True, return_tensors='pt')
output_ids = model.generate(input_ids.to('cuda'))
response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)

print(response)

chatbot.append_response(response)
```

```python
def get_response(chatbot, role, text ):
    messages = chatbot.generate_messages(role = role, text = text)
    for msg in messages:
        msg['role'] = role_dict[msg['role']]
    input_ids = tokenizer.apply_chat_template(conversation=messages, tokenize=True, add_generation_prompt=True, return_tensors='pt')
    output_ids = model.generate(input_ids.to('cuda'))
    response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)
    print(response)
    chatbot.append_response(response)
```

```python
from ChatHaruhi import ChatHaruhi

chatbot = ChatHaruhi( role_name = 'haruhi', max_len_story = 1000 ,llm = "foo")

get_response(chatbot, role='李鲁鲁', text = '春日你好啊，我是新同学李鲁鲁')
```
