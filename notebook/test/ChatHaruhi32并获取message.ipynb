```python
from ChatHaruhi import ChatHaruhi

my_name = "鲁鲁"

messages = []

for name in sugar_names:
    chatbot = ChatHaruhi( role_name = name, llm = get_response )
    message = chatbot.get_message( user = "", text = f"你好我是{my_name}")
    messages.append(message)
```

```python
import pickle
with open('messages.pkl', 'wb') as f:
    pickle.dump(messages, f)
```
